# How the Orioles Lost Too Many 'Should-Have-Won' Games in 2025

**Author:** Matthew Plotnick



## Abstract

As fans, we remember the 2025 Orioles season as a disappointment: a year filled with blown leads, frustrating losses, and games that felt like they should have been wins. But why? How did a talented roster fall short of the playoffs? There are likely many contributing factors, but this analysis focuses on one specific angle: how often the Orioles lost games they were expected to win. By identifying these “should-have-won” games, we can estimate how much those losses impacted their season, and what needs to improve going into 2026.


## Research Question

How do we define a “should-have-won” game?

I began with a rule-based approach, defining these games as ones in which the Orioles:

- Scored 5+ runs  
- Held a 3+ run lead  
- Led after the 6th inning  

However, I expanded the analysis to include a more objective metric: **win probability**. Using data from MLB’s Stats API, any game in which the Orioles reached at least an 80% win probability at any point but still lost was also classified as a “should-have-won” game.


## Pipeline

Using the MLB Stats API, I parsed all **2,430 regular season games** from 2025.

**Key steps:**
- Extracted play-by-play win probability data for each game  
- Removed terminal values (0% and 100%) to avoid skewing results  
- Created opportunity metrics:
  - Games reaching X% win probability (80%, 85%, 90%, 95%)
  - Games leading after 6 innings
  - Games scoring 5+ runs
- Defined failures as losses after reaching those states  
- Calculated **failure rates = failures / opportunities**

This allowed for a consistent comparison of how teams performed when put in strong winning positions.


## League Context

To properly evaluate the Orioles, their performance must be compared to the rest of MLB.

Across these metrics, the Orioles were consistently below league average, though not always extreme outliers. However, teams that ranked poorly in these categories were rarely playoff teams.

The Orioles ranked in the top 10 worst teams in:
- 95% win probability loss rate  
- 80% win probability loss rate  
- Wasted offense (losses when scoring 5+ runs)

## Analysis

The data aligns closely with the eye test.

The Orioles:
- Did not consistently put themselves in position to win  
- Failed to close out games when they did  

Watching it unfold over the course of the season, it often felt like the Orioles were either chasing games early or unable to fully close the door late.

This is likely tied to:
- Inconsistent starting pitching  
- Bullpen instability  
- Injuries (only 3 players avoided injury all season)

But the key takeaway is measurable:

- **95% WP games:** Orioles win rate = **95.2%**; league average = **97.2%** (**5th worst**)  
- **3+ run leads:** Orioles = **87.5%**; league = **90.6%** (**7th worst**)  
- **80% WP games:** Orioles = **82.8%**; league = **86.1%** (**7th worst**)  
- **5+ run games:** Orioles = **74.6%**; league = **79.5%** (**7th worst**)  

Across nearly every metric, the Orioles consistently ranked in the bottom quartile of MLB.

## Comebacks

I also analyzed the opposite scenario: comeback wins.

The Orioles ranked:
- Bottom 10 in nearly every comeback metric  
- **29th (second worst)** in games where they fell to 20% win probability  
- Orioles comeback rate at 20% WP: **7.4%**  
- League average: **12.8%**

This indicates a lack of resilience when trailing.
Unlike the blown leads, these games are easier to forget, but over a full season, the inability to steal wins in these spots slowly adds up.

**Figure 5** shows the distribution of 20% comeback rates across MLB, where the Orioles rank 29th.

While the Orioles had 3 wins from 5% win probability, which were memorable moments, the overall impact was minimal. It amounted to roughly a **1-game swing** over the season. In contrast, blown games represent a much larger gap.

## Conclusion

The Orioles’ 2025 season was not defined by a lack of talent; it was defined by a failure to convert game states into wins.

They struggled in two critical areas:

1. **Protecting leads** — ranking among the worst teams in converting strong win probabilities  
2. **Recovering from deficits** — ranking near the bottom in comeback success  

Even a modest improvement in these areas could have significantly changed the outcome of the season. Converting just a handful of high-probability losses into wins likely would have been enough to push the Orioles into playoff contention.

For 2026, the path forward is clear:
- Improve pitching stability to protect leads  
- Increase offensive consistency to create more opportunities  
- Develop late-game execution, both ahead and behind  

Because in a 162-game season, the difference between missing the playoffs and making a run often comes down to a small number of games, and for the Orioles, those were the games they should have already won.

## Implications for 2026

While this analysis focuses on 2025 outcomes, it naturally raises questions about how the Orioles can improve moving forward.

A new manager may influence late-game decision-making, bullpen usage, and overall game management, all of which directly impact the types of high-leverage situations identified in this analysis. If the Orioles can improve in these areas, particularly in protecting leads and optimizing pitching changes, it could meaningfully reduce their rate of high-probability losses.

However, managerial changes alone are unlikely to fully resolve the issue. The data suggests that both roster-level improvements (particularly in pitching depth and consistency) and strategic execution will be necessary to increase opportunities and convert them into wins.

The Orioles have already made notable roster additions, including Pete Alonso, Taylor Ward, and Ryan Helsley. While these moves may improve specific areas, particularly offensive production and late-inning relief, the results from 2025 suggest that isolated upgrades are unlikely to fully address the broader issue. The team’s challenges were systemic, affecting both opportunity creation and game-state conversion.

Ultimately, the Orioles’ path forward is measurable. If they can move closer to league average in both opportunity creation and game-state conversion, the impact would likely be reflected directly in the standings.

## Appendix

### A1. Data Sources
All data was obtained using the MLB Stats API, including:
- Game schedules and results  
- Play-by-play data  
- Win probability data for each event  

### A2. Definitions

**Win Probability (WP)**  
The probability that a team will win the game at a given moment, based on game state.

**Should-Have-Won Game**  
A loss in which a team reached a specified win probability threshold (80%, 85%, 90%, or 95%) or met rule-based criteria (5+ runs, 3+ run lead, or leading after 6 innings).

**Opportunity Metrics**  
The number of games in which a team reached a specific game state (for example, 90% WP or leading after 6 innings).

**Failure Rate**  
Failures divided by opportunities:

`failure rate = losses / opportunities`

**Comeback Rate**  
Wins divided by opportunities when trailing:

`comeback rate = comeback wins / games in trailing state`

### A3. Methodology Notes
- Terminal win probabilities (0% and 100%) were excluded to avoid bias  
- All metrics were calculated at the team level  
- League averages were computed as the **mean of team-level rates**, ensuring equal weighting across teams  
- Only regular season games (162 per team) were included  

### A4. Limitations
- Win probability does not capture all contextual factors (for example, bullpen strength or matchups)  
- Some late-game states may be disproportionately impactful  
- Results are descriptive and do not establish causation
